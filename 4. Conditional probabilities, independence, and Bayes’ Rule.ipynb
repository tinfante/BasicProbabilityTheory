```python
# Example 4.2.1.
# Suppose I have a jar of marbles, and I choose one
# uniformly at random. Each marble is either red (R),
# green (G), or blue (B). In two-thirds of the
# marbles, these colors are solid (S), while in the
# remainder the colors are patchy. If I tell you
# that the probability of getting a red marble given
# that I have a solid-colored one is 1/2, can you
# compute the probability that my chosen marble is
# solid red?

# We know the following:
# P(R|S) = 1/2
# P(S) = 2/3
#
# According to the product rule, which allows us to
# compute joint probabilities from conditional
# probabilities, we have:
# P(R|S) = P(R∩S)/P(S)
# P(R∩S) = P(R|S)*P(S)
# P(R∩S) = 1/2 * 2/3 = 2/6 = 1/3
```

```python
# Example 4.2.2.
# Now suppose the probability of the marble being
# blue given that it’s patchy is 1/5. What is the
# probability of getting a patchy blue marble?

# We know:
# P(B|P) = 1/5
# P(P) = 1/3
#
# According to the product rule:
# P(B|P) = P(B∩P)/P(P)
# P(B∩P) = P(B|P)*P(P)
# P(B∩P) = 1/5 * 1/3 = 1/15
```

```python
# Example 4.3.1.
# Using the same jar of marbles as in Examples 4.2.1
# and 4.2.2, I give you one more piece of information:
# the probability of a green marble given a patchy
# marble is 2/5. What’s the probability of a red
# marble given a patchy marble?

# We have the following information:
#
# P(B|P) = 1/5
# P(G|P) = 2/5
#
# We can know that P(R|P) is 2/5 because with P(B|P)
# and P(G|P) they make a conditional distribution,
# so they should add up to one.
```

```python
# Example 4.6.1.
# Suppose I have some playing cards in my hand from
# a standard deck, and it turns out that 1/4 of them
# are red cards (hearts or diamonds) and 3/8 are
# face cards (jack, queen, or king). Also, 1/2 of
# the red cards are face cards. If I choose one of
# the cards in my hand uniformly at random, and get
# a face card, what’s the probability that it is
# also a red card?

# Bayes rule:
# P(A|B) = (P(B|A) * P(A)) / P(B)
# 
# We have the following information:
#     P(R) = 1/4
#     P(F) = 3/8
#     P(F|R) = 1/2
# and we need P(R|F)
#
# P(R|F) = (P(F|R) * P(R)) / P(F)
# P(R|F) = (1/2 * 1/4) / 3/8
# P(R|F) = 1/8 / 3/8 = 1/3
```

```python
# Example 4.6.2. In the general population, 1/1000
# infants have a particular genetic marker that is
# linked to cancer in later life. There is a test
# for this marker, but it isn’t perfect. If an infant
# has the genetic marker, the test will always show a
# positive result. But it also has a false positive
# rate of 1% (that is, 1% of infants who do not have
# the marker will still show a positive test result).
# If we test a particular infant and the test comes
# out positive, what’s the probability that infant
# actually has the genetic marker?

# We have this information,
#     P(M) = 1/1000
#     P(P|M) = 1
#     P(P|¬M) = 1/100
# and asked to compute P(M|P). Lets start with
# Bayes' Rule.
#
# P(M|P) = (P(P|M) * P(M)) / P(P)
# P(M|P) = (1 * 1/1000) / P(P)
# P(M|P) = 1/1000 / P(P)
#
# We need to get the value of the denonimator
# P(P), which we can do with the law of total
# probability.
#
# P(P) = P(P∩M) + P(P∩¬M)
#
# Now lets use the product rule
# P(P) = P(P|M) * P(M) + P(P|¬M) * P(¬M)
#
# We can get P(¬M) by the complement, so
# P(¬M) = 1 - P(M) = 1 - 1/1000 = 999/1000
#
# P(P) = 1 * 1/1000 + 1/100 * 999/1000
# P(P) = 1/1000 + 999/100000
# P(P) = 100/100000 + 999/100000
# P(P) = 1099/100000
#
# We have the denominator. So lets calculate
# P(M|P) = 1/1000 / 1099/100000
print('%.2f' % ((1/1000) / (1099/100000)))
```
